# Rejection Sampling 


```python
from scipy.stats import norm 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
%config InlineBackend.figure_format="retina" 
```

```python
def p(x):
    return norm.pdf(x, loc=30, scale=10) + norm.pdf(x, loc=80, scale=20) 

def q(x):
    return norm.pdf(x, loc=50, scale=30) 
```

```python
x = np.arange(-50,150) 
fig, ax = plt.subplots(figsize=(12,6)) 
ax.plot(x, p(x), label="$p(x)$")
ax.plot(x, q(x), label="$q(x)$") 
plt.legend() 
plt.show()
```

```python
k = max(p(x)/ q(x))
```

```python
x = np.arange(-50,150) 
fig, ax = plt.subplots(figsize=(12,6)) 
ax.plot(x, p(x), label="$p(x)$")
ax.plot(x, k*q(x), label="$k \cdot q(x)$") 
plt.legend() 
plt.show()
```

```python
def sample(size):
    xs = np.random.normal(50,30,size=size) 
    cs = np.random.uniform(0,1,size=size) 
    mask = (p(xs) / (k*q(xs))) > cs 

    return xs[mask]
```

```python
samples = sample(10000) 
sns.displot(samples) 
plt.show()
```
